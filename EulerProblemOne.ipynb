```python
#If we list all the natural numbers below 10 that are multiples of  3 or 5, 
#we get  3,5,6,9 . The sum of these multiples is 23. Find the sum of all the multiples of 3 or 5 below 1000
```

```python
def multiplesum(maximum, factor1, factor2):
    '''Returns the sum of all numbers which are multiples of factor1 or factor2, up to maximum.'''
    running_sum = 0
    for i in range(maximum):
        if i % factor1 == 0 or i % factor2 == 0:
            running_sum += i
    return running_sum
```

```python
sum_fortest = multiplesum(1000, 6, 7)
print(sum_fortest)
```

```python
def multiplesum_nfactors_np(maximum, factors):
    '''Returns the sum of all numbers which are multiple of any factor in the list factors, using numpy.'''
    running_sum = 0
    for i in range(maximum):
        # i % factors returns a numpy array factors where each entry is i % factors in place of factors 
        if np.any(i % np.asarray(factors) == 0):
            #print(i % np.asarray(factors))
            running_sum += i
    return running_sum
```

```python
sum_fortest_3 = multiplesum_nfactors_np(1000, [6, 7])
print(sum_fortest_3)
```
