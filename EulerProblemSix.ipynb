```python
'''PROBLEM STATEMENT:
Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.
'''
```

```python
def sumofsquares(maximum):
    '''Returns sum of squares of natural numbers up to and including maximum.'''
    total = 0
    for i in range(1,maximum+1):
        total += i ** 2
    return total
```

```python
sumofsquares(10)
```

```python
def squareofsum(maximum):
    '''Returns the square of the sum of natural numbers up to and including maximum.'''
    total = sum(range(maximum+1))
    return total ** 2
```

```python
squareofsum(10)
```

```python
def square_diff(maximum):
    '''Returns difference between square of sum and sum of squares.'''
    return squareofsum(maximum)-sumofsquares(maximum)
```

```python
square_diff(10)
```

```python
square_diff(100)
```
